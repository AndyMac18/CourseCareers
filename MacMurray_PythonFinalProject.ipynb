```python
import seaborn as sns
import pandas as pd
```

```python
mpg = sns.load_dataset("mpg")
```

```python
mpg.head()
```

```python
#Create a visual showing the relationship between 
#horsepower and fuel economy.
sns.relplot(
    data=mpg,
    x="horsepower", y="mpg")
```

```python
#Create a table summarizing fuel economy by country of origin.
#Create a visual showing how fuel economy has changed over time
sns.catplot(data=mpg,
            x="model_year", y="mpg",
            hue="origin",
            kind="bar", errorbar=None)
```

```python
#Create two other visuals or tables related to fuel economy.
#1. Create a plot showing the relationship between vehicle weight and fuel economy.
sns.relplot(data=mpg,
            x="weight",
            y="mpg")
```

```python
#Create two other visuals or tables related to fuel economy.
#2. Create a chart showing the average MPG based on the amount of engine cylinders, sorted by origin.
sns.catplot(data=mpg,
            x="cylinders",
            y="mpg",
            hue="origin",
            kind="bar",
            errorbar=None)
```

```python
#Comments & Observations
#The cars in this database have an inverse relationship between horsepower and fuel economy.
    #In general, the higher a car's horsepower, the worse its fuel economy.
#The average fuel economy for cars increased between 1970-1982.
    #Cars with US origin had the worst average fuel economy every year.
    #Japanese and European cars traded the top spot for best average fuel economy several times.
#There is also an inverse relationship between weight and fuel economy.
    #In general, the higher a car's weight, the worse the fuel economy.
#In general, cars with four cylinder engines have the best fuel economy. Cars with eight cylinders have the worst.
    #In this database, the only cars with eight-cylinder engines are of American origin.
    #The only cars with five-cylinder engines are of European origin.
    #The only cars with three-cylinder engines are of Japanese origin.
    #In four & six cylinder engine cars, Japanese cars on average have better fuel economy than their European and American counterparts.
```
